```python
x=data.drop('target1',axis=1)
y=data.target1
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=10)
k = pca.fit_transform(x)
final_data=pd.DataFrame(data=k,columns=['column1','column2','column3','column4','column5','column6','column7','column8',
                                        'column9','column10'])
```

```python
from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
for i in final_data.columns:
    final_data[i]=scale.fit_transform(final_data[[i]])
```

```python
final_data.hist(bins=25,figsize=(20,20))
```

```python
final_data
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(final_data,y,test_size=0.33,random_state=42)
```

```python
import matplotlib.pyplot as plt
def mapper(x,y):
    plt.scatter(range(len(x)),x)
    plt.title("predicted")
    plt.show()
    
    plt.scatter(range(len(y)),y)
    plt.title("test")
    plt.show()
    
```

```python
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier(max_depth=3)
trf=tree.fit(xtrain,ytrain)
pre = trf.predict(xtest)

from sklearn.metrics import accuracy_score
score=accuracy_score(pre,ytest)
print(score*100)
mapper(pre,ytest)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest= RandomForestClassifier(max_depth=2,n_estimators=1000)
forfit=forest.fit(xtrain,ytrain)
pre=forfit.predict(xtest)
score = accuracy_score(pre,ytest)
print(score*100)
mapper(pre,ytest)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kne = KNeighborsClassifier(n_neighbors=3)
knn_fit=kne.fit(xtrain,ytrain)
pre=knn_fit.predict(xtest)
score = accuracy_score(pre,ytest)
print(score*100)
plt.scatter(range(len(pre)), pre, label='Predicted')
plt.scatter(range(len(ytest)), ytest, label='Test')
plt.xlabel('Data Points')
plt.ylabel('Values')
plt.title('Comparison of Predicted and Test Values')
plt.legend()
plt.show()
```

```python
from sklearn.svm import SVC
svc= SVC(kernel='sigmoid')
sv_fit = svc.fit(xtrain,ytrain)
svc_prd = sv_fit.predict(xtest)
score =accuracy_score(svc_prd,ytest)
print(score*100)
mapper(svc_prd,ytest)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr_fit = lr.fit(xtrain,ytrain)
lr_pre= lr_fit.predict(xtest)
score = accuracy_score(lr_pre,ytest)
print(score*100)
mapper(lr_pre,ytest)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb_fit=gnb.fit(xtrain, ytrain)
gnb_pre=gnb_fit.predict(xtest)
score  = accuracy_score(gnb_pre,ytest)
print(score*100)
mapper(gnb_pre,ytest)
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='lbfgs', alpha=1e-5,random_state=2)
clf_fit = clf.fit(xtrain,ytrain)
clf_pre= clf_fit.predict(xtest)
score = accuracy_score(clf_pre,ytest)
print(score*100)
mapper(clf_pre,ytest)
```
